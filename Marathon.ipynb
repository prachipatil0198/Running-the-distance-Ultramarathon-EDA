```python
df3.groupby(['race_length','athlete_gender'])['athlete_average_speed'].mean()
```

```python
#What age groups are the best in the 50m Race
```

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False).query('count>19').head(10)
```

```python
#Seasons for the data -> Slower in summer than winter

#spring 3-5
#summer 6-8
#fall 9-11
#winter 12-2

#spliting between two decimals
```

```python
df3['race_month']=df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x>11 else 'Fall'if x>8 else'Summer' if x>5 else'Winter')
```
